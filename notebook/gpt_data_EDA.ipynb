# Lấy dữ liệu theo type của các cột Score_for_answers

```python
df.dropna(inplace=True)
```

```python
type_columns = [col for col in columns if "Score_for_answers" in col]
```

```python
def convert_str_to_list(str_list):
    if isinstance(str_list, list):
        return str_list

    try:
        # Remove brackets and split by comma
        str_list = str_list.strip('[]')
        # Convert each element to integer
        num_list = [int(x.strip()) for x in str_list.split(',')]
        return num_list
    except:
        return []
```

```python
for col in type_columns:
    col_type_name = col.replace("Score_for_answers", f"Score_answers_median")
    df[col] = df[col].apply(convert_str_to_list)
    df[col_type_name] = df[col].apply(lambda x: np.median(x))
```

### Xử lý conflict trong cột vqac_reason_depth_response

```python
df['vqac_reason_depth_response'] = pd.to_numeric(df['vqac_reason_depth_response'], errors='coerce')
```

```python
df.dropna(inplace=True)
```

```python
df['index'] = df['index'].astype(str)
```
